```python
data_final["description"] = data_final["description"].str.strip()
```

```python
data_final.dropna(axis=0, subset=['uuid'], inplace=True)
```

```python
data_final["uuid"] = data_final["uuid"].astype("str")
```

# 轉變為購物籃數據集

```python
basket = data_final.pivot_table(columns="name",index="uuid",values = "quantity",aggfunc=np.sum).fillna(0)
```

```python
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
```

```python
basket_sets.head()
```

# 關聯性分析

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.00001 ,use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.01)
```
